```python
import pandas as pd
import numpy as np

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/OLYMPICS_athlete_events.csv")

#1
numberOfYears = len(pd.unique(data["Year"]))
print("There are", numberOfYears, "different years represented in the data")

numberOfCountries = len(pd.unique(data["NOC"]))
print("There are", numberOfCountries, "different countries represented in the data")
print()



#2
usaGold = 0
usaSilver = 0
usaBronze = 0
mexGold = 0
mexSilver = 0
mexBronze = 0
canGold = 0
canSilver = 0
canBronze = 0

for index, row in data.iterrows():
  if (row["NOC"] == "USA"):
    if(row["Medal"] == "Gold"):
      usaGold += 1
    elif (row["Medal"] == "Silver"):
      usaSilver += 1
    elif (row["Medal"] == "Bronze"):
      usaBronze += 1
    else:
      continue
  elif (row["NOC"] == "MEX"):
    if(row["Medal"] == "Gold"):
      mexGold += 1
    elif (row["Medal"] == "Silver"):
      mexSilver += 1
    elif (row["Medal"] == "Bronze"):
      mexBronze += 1
    else:
      continue
  elif (row["NOC"] == "CAN"):
    if(row["Medal"] == "Gold"):
      canGold += 1
    elif (row["Medal"] == "Silver"):
      canSilver += 1
    elif (row["Medal"] == "Bronze"):
      canBronze += 1
    else:
      continue
  

print("The number of USA gold medals is", usaGold)
print("The number of USA silver medals is", usaSilver)
print("The number of USA bronze medals is", usaBronze)
print("The number of MEX gold medals is", mexGold)
print("The number of MEX silver medals is", mexSilver)
print("The number of MEX bronze medals is", mexBronze)
print("The number of CAN gold medals is", canGold)
print("The number of CAN silver medals is", canSilver)
print("The number of CAN bronze medals is", canBronze)
print()


#3
eventData = data[["Season", "Year", "Event"]]

summerData = eventData[(eventData["Season"] == "Summer") & (eventData["Year"] > 2000)]
winterData = eventData[(eventData["Season"] == "Winter") & (eventData["Year"] > 2000)]

summerEvents = len(pd.unique(summerData["Event"]))
winterEvents = len(pd.unique(winterData["Event"]))

print("The number of different Summer Olympic events taking place after 2000 is", summerEvents)
print("The number of different Winter Olympic events taking place after 2000 is", winterEvents)
print()


#Bonus Question
#What is the average BMI of all competitors in the dataset?

bmi = data["Weight"] / (.01 * data["Height"])**2
avgBMI = bmi.mean()

print("The average BMI of all competitors in the dataset is", avgBMI)
```

```python
import pandas as pd
import numpy as np

olympic_data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/OLYMPICS_athlete_events.csv")

########4. Compare medal counts across all 3 (Gold, Silver and Bronze) for summer olympics for China
#and the USA for the year 2016.########
summer2016_data = olympic_data[(olympic_data["Year"] == 2016) & (olympic_data["Season"] == "Summer")]

china_gold = 0
china_silver = 0
china_bronze = 0 

usa_gold = 0
usa_silver = 0
usa_bronze = 0

for index,row in summer2016_data.iterrows():
  if row["NOC"] == "CHN":
    if row["Medal"] == "Gold":
      china_gold += 1
    elif row["Medal"] == "Silver":
      china_silver += 1
    elif row["Medal"] == "Bronze":
      china_bronze += 1
    else:
      continue
  elif row["NOC"] == "USA":
    if row["Medal"] == "Gold":
      usa_gold += 1
    elif row["Medal"] == "Silver":
      usa_silver += 1
    elif row["Medal"] == "Bronze":
      usa_bronze += 1
    else:
      continue
  else:
      continue

print("China's medals in 2016 Summer Olympics:", china_gold, "Gold, ", china_silver, "Silver, ", china_bronze, "Bronze.")
print("USA's medals in 2016 Summer Olympics:", usa_gold, "Gold, ", usa_silver, "Silver, ", usa_bronze, "Bronze.")
print()



#########5. How many medals did the US win in Summer vs Winter games after the year 2000.########

USA_medals_data = olympic_data[(olympic_data["Year"] > 2000) & (olympic_data["NOC"] == "USA")]
Summer_medals= 0
Winter_medals= 0

for index,row in USA_medals_data.iterrows():
  if row["Season"] == "Summer":
    if row["Medal"] == "Gold" or row["Medal"] =="Silver" or row["Medal"] =="Bronze":
      Summer_medals += 1
    else:
      continue
  elif row["Season"] == "Winter":
    if row["Medal"] == "Gold" or row["Medal"] =="Silver" or row["Medal"] =="Bronze":
      Winter_medals += 1
    else:
      continue
  else:
      continue 
print("The US won",Summer_medals,"medals in the Summer games after the year 2000")
print("The US won",Winter_medals,"medals in the Winter games after the year 2000") 
print()

##########6. What is the average age of male vs female contestants for the USA and China in games after
#the year 2000.###########
usa_contestants = olympic_data[(olympic_data["Year"] > 2000) & (olympic_data["NOC"] == "USA")]
usa_age_avg = usa_contestants.groupby(["Sex"]).mean()

print("The average age of USA contestants after 2000:", usa_age_avg["Age"])

china_contestants = olympic_data[(olympic_data["Year"] > 2000) & (olympic_data["NOC"] == "CHN")]
china_age_avg = china_contestants.groupby(["Sex"]).mean()

print("The average age of a China contestants after 2000:", china_age_avg["Age"])
print()

####custom question- compare the number of medals won by the US in 2016 and 2000?#####
US2016_medals = olympic_data[(olympic_data["Year"] == 2016) & (olympic_data["NOC"] == "USA")]
US2016_medals = US2016_medals[US2016_medals['Medal'].notna()]
print('In 2016, the US recieved',US2016_medals['Medal'].count(),'medals')

US2000_medals = olympic_data[(olympic_data["Year"] == 2000) & (olympic_data["NOC"] == "USA") & (olympic_data["Medal"].notna())]
print('In 2000, the US recieved',US2000_medals['Medal'].count(),'medals')
print('The US recieved',US2016_medals['Medal'].count()-US2000_medals['Medal'].count(), 'more medals in 2016 than in 2000')



```

#8. Compare the countries of Denmark, Norway and Sweden in medal counts across the entire data set.

```python
import numpy as np
import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/OLYMPICS_athlete_events.csv")

#Denmark
Den = data["Team"] == "Denmark"
data["Den"]= Den

den_Medal = data[["Den","Medal"]].groupby("Den").count()
print("Denmark's medal counts is:")
print(den_Medal)
print()

#Norway
Nor = data["Team"] == "Norway"
data["Nor"] = Nor

nor_Medal = data[["Nor","Medal"]].groupby("Nor").count()
print("Norway's medal counts is:")
print(nor_Medal)
print()

#Sweden
Swe = data["Team"] == "Sweden"
data["Swe"] = Swe

swe_Medal = data[["Swe","Medal"]].groupby("Swe").count()
print("Sweden's medal counts is:")
print(swe_Medal)
print()
```
